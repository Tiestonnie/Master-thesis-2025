```python
import shap
import numpy as np

# Load the data
data = np.load("C:/Users/31610/Documents/UvA/Master thesis/Data/SHAP/shap_values_lead_14_seed_133_exp_1_exp_100.h5_new.npz", allow_pickle=True)
print(data)
shap_values = data['shap_values']  # Adjust key based on your file
features = data['feature_names']        # Adjust key if feature names are included

# Example: Summary plot (requires a model and feature data)
shap.summary_plot(shap_values, features)
mean_impact = np.abs(shap_values).mean(axis=0)
```

```python
import matplotlib.pyplot as plt
# Compute mean absolute SHAP impact
mean_impact = np.abs(shap_values).mean(axis=0)

# Create bar chart
plt.bar(features[:mean_impact.shape[0]], mean_impact)
plt.xlabel('Features')
plt.ylabel('Mean Absolute SHAP Impact')
plt.title('Mean SHAP Impact per Feature')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
